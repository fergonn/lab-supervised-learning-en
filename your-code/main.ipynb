Your features will be:

+ URL: it is the anonymous identification of the URL analyzed in the study
+ URL_LENGTH: it is the number of characters in the URL
+ NUMBER_SPECIAL_CHARACTERS: it is number of special characters identified in the URL, such as, “/”, “%”, “#”, “&”, “. “, “=”
+ CHARSET: it is a categorical value and its meaning is the character encoding standard (also called character set).
+ SERVER: it is a categorical value and its meaning is the operative system of the server got from the packet response.

+ WHOIS_COUNTRY: it is a categorical variable, its values are the countries we got from the server response (specifically, our script used the API of Whois).
+ WHOIS_STATEPRO: it is a categorical variable, its values are the states we got from the server response (specifically, our script used the API of Whois).
+ WHOIS_REGDATE: Whois provides the server registration date, so, this variable has date values with format DD/MM/YYY HH:MM
+ WHOIS_UPDATED_DATE: Through the Whois we got the last update date from the server analyzed
+ TCP_CONVERSATION_EXCHANGE: This variable is the number of TCP packets exchanged between the server and our honeypot client
+ DIST_REMOTE_TCP_PORT: it is the number of the ports detected and different to TCP
+ REMOTE_IPS: this variable has the total number of IPs connected to the honeypot
+ APP_BYTES: this is the number of bytes transfered
+ SOURCE_APP_PACKETS: packets sent from the honeypot to the server
+ REMOTE_APP_PACKETS: packets received from the server
+ APP_PACKETS: this is the total number of IP packets generated during the communication between the honeypot and the server
+ DNS_QUERY_TIMES: this is the number of DNS packets generated during the communication between the honeypot and the server
+ TYPE: this is a categorical variable, its values represent the type of web page analyzed, specifically, 1 is for malicious websites and 0 is for benign websites

```python
# Select only numeric columns for correlation analysis
numeric_df = websites.select_dtypes(include=['float64', 'int64'])

# Create the correlation matrix
corr_matrix = numeric_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Create a mask to hide the upper triangle (since it's a mirror of the lower)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Draw the heatmap
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap='RdBu_r', center=0)
plt.title('Correlation Heatmap of Website Features')
plt.show()
```

```python
import numpy as np

# 1. Compute the absolute correlation matrix
# We use .abs() because both 0.95 and -0.95 indicate strong collinearity
corr_matrix = websites.select_dtypes(include=[np.number]).corr().abs()

# 2. Select the upper triangle of the correlation matrix 
# (This prevents comparing a column to itself or seeing the same pair twice)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# 3. Find index of columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

print(f"Number of highly correlated columns to consider dropping: {len(to_drop)}")
print(f"Suggested columns to drop: {to_drop}")

# 4. Detailed breakdown of the pairs
print("\nDetailed Correlation Pairs (> 0.95):")
for column in upper.columns:
    high_corr = upper[column][upper[column] > 0.95]
    for index, value in high_corr.items():
        print(f"- {index} and {column} are {value:.4f} correlated.")
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
def clean_server(server_string):
    # Ensure we are working with a string and handle NaNs if any remain
    server_string = str(server_string).lower()
    
    if 'microsoft' in server_string:
        return 'Microsoft'
    elif 'apache' in server_string:
        return 'Apache'
    elif 'nginx' in server_string:
        return 'nginx'
    else:
        return 'Other'

# Apply the function to the SERVER column
websites['SERVER'] = websites['SERVER'].apply(clean_server)
```

#### Our second algorithm is is DecisionTreeClassifier

Though is it not required, we will fit a model using the training data and then test the performance of the model using the testing data. Start by loading `DecisionTreeClassifier` from scikit-learn and then initializing and fitting the model. We'll start off with a model where max_depth=3

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the model with a max depth of 3
# random_state ensures your results are reproducible
dt_model = DecisionTreeClassifier(max_depth=3, random_state=42)

# Fit the model using your training data
dt_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Make predictions
y_pred_dt = dt_model.predict(X_test)

# Calculate metrics
dt_accuracy = accuracy_score(y_test, y_pred_dt)
dt_cm = confusion_matrix(y_test, y_pred_dt)

print(f"Decision Tree Accuracy: {dt_accuracy:.4f}")
print("\nDecision Tree Confusion Matrix:")
print(dt_cm)
```

To test your model, compute the predicted probabilities, decide 0 or 1 using a threshold of 0.5 and print the confusion matrix as well as the accuracy score (on the test set!)

```python
# Get predicted probabilities for the 'Malicious' class (column 1)
y_probs_dt = dt_model.predict_proba(X_test)[:, 1]

# Apply the 0.5 threshold
# If probability >= 0.5, label as 1 (Malicious), else 0 (Benign)
y_pred_threshold = (y_probs_dt >= 0.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# Calculate Accuracy
dt_accuracy = accuracy_score(y_test, y_pred_threshold)

# Generate Confusion Matrix
dt_cm = confusion_matrix(y_test, y_pred_threshold)

print(f"Decision Tree (Depth=3) Accuracy: {dt_accuracy:.4f}")
print("-" * 30)
print("Confusion Matrix:")
print(dt_cm)
```
